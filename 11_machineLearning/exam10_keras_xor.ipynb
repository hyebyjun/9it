### 인공지능의 역사

```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
```

```python
training_data = np.array(
    [[0, 0], [0, 1], [1, 0], [1, 1]], 'float32')
target_data = np.array([[0], [1], [1], [0]])
```

```python
# 케라스에서 모델 만드는 방법~

model = Sequential()
model.add(Dense(32, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mse', optimizer='adam', metrics=['binary_accuracy'])
print(model.summary())
```

```python
fit_hist = model.fit(training_data, target_data, epochs=500, verbose=1)
# 입력데이터, 예측값과 비교될 목표데이터, 학습횟수)
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
plt.plot(fit_hist.history['binary_accuracy'])
plt.show()  # 할때마다 그래프 다르다?
```

```python
inp = list(map(int, input('논리값을 입력하세요.').split()))
qwe = np.array(inp)  # input값을 배열로 바꿔서
print('입력값 :', qwe)
qwe = qwe.reshape(-1, 2)  # qwe값이 총 몇 개가 있든 2개씩 쪼갤거고 묶음단위 맞든 안맞든 -1 <== 알아서해라
print('reshape :', qwe)
print('결과값 :', model.predict(qwe)[0][0].round())
```
