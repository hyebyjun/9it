#plt.figure 생성 후 add_axes해서 ax.plot으로 한번에 여러개의 선그래프 만들기

```python
point= 50
X= np.linspace(-5, 5, point)  #-5부터 5까지를 50개로 나눈 넘파이 객체 생성
fig= plt.figure(figsize=(8,5))
ax= fig.add_axes([0,0,1,1])

n=50
x=np.random.rand(n) * 10 - 5           #랜덤하게 n개의 데이터를 만듦. 0에서 1 사이 (0포함 1 미포함). 그거에 10 곱했으니 0~10이고 -5니까 0~5 수 n(50)개.
y= np.random.rand(n)*35 - 10      #0~25
custom_colors=np.random.rand(n)   #0~1사이의 50개의 랜덤값.
area= (30*np.random.rand(n))**2   #점의 크기.  scatter는 점그래프잖아. 거기에 s는 size거든, 점의 사이즈. 점 사이즈를 랜덤하게 한다는 거고. 색도 랜덤. 알파는 투명도. 투명도는 0.5로 고정.
ax.scatter(x,y,s=area, c=custom_colors, alpha=0.5)
#x에 x, y에 y, 점사이즈는 area, 색은 custom_colors (다 넘파이객체다).


ax.plot(X, 2*X+3, 'rD', X, np.sin(X)*10, 'b^', X, X**2, 'gs')   # 3개의 선 한번에 그린거임. 여기에 계속 추가할 수 있음.
plt.axhline(color='black', linewidth=1)
plt.axvline(color='black', linewidth=1)
plt.show() 
```

# scatter 그래프로 표현하는 sigmoid 함수

```python
import math
def sigmoid(x):
  sig = 1 / (1 + math.exp(-x))
  return sig

point = 100
X = np.linspace(-10, 10, point)   #linspace로 시작값, 끝값, 나누는 개수
Y = list(map(sigmoid, X))

fig = plt.figure(figsize = (8,5))
ax = fig.add_axes([0, 0, 1, 1])
np.random.seed(100)
N = 50
n = 50
x = np.random.rand(n) * 20 - 10 
y = np.random.rand(n) * 1.4 - 0.2 
custom_colors = np.random.rand(n) # 색상도 random (숫자로 표현 가능)
area = (30* np.random.rand(n)) ** 2 # 0~ 900 의 점 크기
ax.scatter(x,y, s = area, c = custom_colors, alpha = 0.5) # s  = 점의 크기, alpha = 점의 투명도
ax.plot(X, Y, color = 'r', linewidth = 2)

ax.fill_between(X, list(map(lambda Y:Y+0.25, Y)),
                list(map(lambda Y:Y-0.25, Y)), 
                facecolor = 'r', edgecolor = 'None', alpha = 0.15)

plt.axhline(color = 'k', linewidth = 1)
plt.axvline(c = 'k', linewidth = 1)
ax.set_title('Sigmoid')
```

# 파이그래프와 정규표현식

```python
labels= 'Frogs','Hogs', 'Hogs', 'Logs'
sizes = [15, 30, 45, 10]
explode= (0, 0.1, 0, 0)  #두번째꺼가 0.1만큼 튀어나옴

fig1, ax1 = plt.subplots(figsize=(15,7))

#파이그래프 만들 때  wedges, texts, autotexts넣은 후 조절한다.
wedges, texts, autotexts = ax1.pie(sizes, labels=labels, explode=explode, shadow=True, startangle= -90, autopct= '%1.3f%%' )  
 #소수점아래 3째자리까지 보여주겠다는. 여기서 %는 포맷문자. %뒤는 출력포맷에 대한 설정이라는 거임. 1.2f는 숫자 출력형태를 말함. f는 float을 의미함.
#1.3f 1은, 만약 없으면 1개라도 0을 1개라도 적으라는 거.   그럼 뒤에 퍼센트를 찍고 싶으면, 퍼센트 두번.하면, 하나는 포맷문자고 하나는 퍼센트를 찍어라,라는 의미가 됨.

ax1.axis('equal')  #이거하면 가운데로 감.


wedges[0].set(hatch='///', alpha=0.3)
# wedges[1].set_radius(1.1)
# wedges[3].set_theta1(90)
# wedges[3].set_theta2(90)   #세타 1은 시작, 세타2는 끝나는 각도


plt.show()
```

```python
size= 0.3
vals1= [40,35,25]     #바깥 부채꼴
vals2 = [11,29,12,23,12,13]  #안쪽부채꼴

cmap= plt.get_cmap('tab20c')  #tab20c라는 칼라맵을 이용하는 거.
outer_colors = cmap([0,4,8])    #만든 칼라맵 객체를 이용해서 칼라리스트(?)를 만든다
inner_colors = cmap([1,2,5,6,9,10])

fig = plt.figure(figsize=(8,5))
ax= fig.add_axes([0,0,1,1])  #fig에 위치설정하여 ax생성.

ax.pie(vals1, radius=1 , colors= outer_colors, labels=['male','female','pet'],
       autopct='%1.1f%%', pctdistance=0.83, textprops={'color':'black'}, wedgeprops= {'width':size, 'edgecolor':'w', 'linewidth':5},
       startangle=0, counterclock=True
       )   #radius=1이 디폴트                #pctdistance는 퍼센트값의 위치를 의미함.         #wedge는 하나하나의 부채꼴을 의미함.   #startangle은 0이 디폴트.
       #rotatelabels=False가 디폴트. 글자도 휘게할 건지.



ax.pie(vals2, radius=1-size , colors= inner_colors, 
       autopct='%1.1f%%', pctdistance=0.8, textprops={'color':'w'}, wedgeprops= {'width':size, 'edgecolor':'w', 'linewidth':5},
       startangle=0, counterclock=True
       ) 

plt.show()
```

```python
cmaps={}

gradient = np.linspace(0, 1, 256)
gradient= np.vstack((gradient, gradient))

def plot_color_gradients(category, cmap_list): #인자로 카테고리와 cmap_list.
  nrows = len(cmap_list)
  figh = 0.35 + 0.15 + (nrows+(nrows-1)*0.1) * 0.22
  fig, axs= plt.subplots(nrows=nrows+1, figsize=(6.4, figh))       
  fig.subplots_adjust(top=1-0.35/figh, bottom=0.15/figh, left=0.2, right=0.99)
  axs[0].set_title(f'{category} colormaps', fontsize=14)
  for ax, name in zip(axs, cmap_list):
    ax.imshow(gradient, aspect='auto', cmap=plt.get_cmap(name))
    ax.text(-0.01, 0.5, name, va='center', ha='right', fontsize=10, transform=ax.transAxes)
  for ax in axs:
    ax.set_axis_off()
  cmaps[category] = cmap_list
```

```python
plot_color_gradients('Perceptually Uniform Sequential', 
                     ['viridis','plasma','inferno','magma','cividis']
                     )
```

# 컬럼값 분류해보자

```python
titanic= sns.load_dataset('titanic')
titanic_age = titanic[['age', 'survived']].dropna()   #나이에 따른 생존율 보려고 새 df만듦.
print(titanic_age.head())
print(titanic_age.info())
print(titanic_age.age.max())
```

# 컬럼값 분류를 위해 각 컬럼값에 apply로 함수 적용

```python
def age_band(num):
  for i in range(1, 10):
    if num < 10*i: 
      return f'under {i*10}'
titanic_age['age_band'] = titanic_age['age'].apply(age_band)    #age_band컬럼 만듦
print(titanic_age.head())
```

# 그룹핑 후 그룹마다의 새로운 컬럼값 생성

```python
titanic_age = titanic_age[[
                           'age_band','survived']].groupby('age_band')[
                          'survived'].value_counts().sort_index().unstack().fillna(0)
#titanic_age df 중 age_band와 survived 컬럼만 가져온 후, age_band로 그룹핑한 결과 중 survived 컬럼의 카운트를 인덱스별로 정리하고 결측값엔 0을 넣어라.


titanic_age['Survival rate'] = titanic_age[1]/(titanic_age[0]+titanic_age[1]) * 100        
#survival rate라는 컬럼을 만드는데, 1컬럼을(생존자 카운팅을)   0+1(총 탑승자수 카운팅으로)나눈 것x100의 비율로 값을 계산한다.
print(titanic_age)          
print(type(titanic_age))
```

# 구간별 분포를 나타내는 히스토그램 

```python
#히스토그램 - 구간별 분표를 보려는 거임. 도수분포표. 도수 = 빈도수. 빈도수분포표.


fig, ax= plt.subplots(1,2, figsize=(16,5))  #맷플롯 객체 안에 1행2열로 그리드 나눔.

#bar1
ax[0].bar(titanic_age.index, titanic_age['Survival rate'], color='gray')
ax[0].set_title('Age Band & Survival Rate(Before)' )

#bar2
from matplotlib.ticker import FuncFormatter

color_map= ['gray']*9  #리스트를 곱하면 반복
color_map[0]=color_map[8] = '#3caea3'   #그중 강조하고 싶은 2개만 다른 색을 줬음.
ax[1].bar(titanic_age.index, 
          titanic_age['Survival rate'], color=color_map, edgecolor= 'k',
          linewidth= 1.2, alpha=0.7
          )
ax[1].set_title(
    'Age Band & Survival Rate(After)'
)

for i, rate in enumerate(titanic_age['Survival rate']):
  ax[1].annotate(f'{rate:.02f}%', xy=(i, rate+2),
  va='center', ha='center',
  fontweight='bold', color='#383838')

ax[1].yaxis.set_major_formatter(
    FuncFormatter(lambda y, _: f'{y:}%'))
plt.suptitle('survival rates for young and old')  #전체 피겨의 타이틀
plt.subplots_adjust(left=0.5, right=0.8)
plt.tight_layout()

plt.show()
```
