### 기존 방식

```python
def celsius_to_fahrenheit(x):
    return x*1.8 + 32
```

```python
celsius_value = int(input())
print('섭씨 온도 :',celsius_value,'--> 화씨 온도 :', celsius_to_fahrenheit(celsius_value))
```

### 머신러닝 방식

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, InputLayer
import numpy as np
import matplotlib.pyplot as plt
```

```python
data_C = np.array(range(0, 100))
data_F = celsius_to_fahrenheit(data_C)
print(data_C)
print(data_F)
```

```python
model = Sequential()
model.add(InputLayer(input_shape=(1,)))
model.add(Dense(1))
model.compile(loss='mse', optimizer='rmsprop')
model.summary()
```

```python
model.save('./before_learning.h5')
```

```python
scaled_data_C = data_C / 100
scaled_data_F = data_F / 100
print(scaled_data_C)
print(scaled_data_F)
```

```python
print(model.predict([0.01]))
```

```python
fit_hist = model.fit(scaled_data_C, scaled_data_F, epochs=1000)
# (A, B, C) A를 입력하면 B를 출력, C번 하세요
```

```python
model.save('./after_learning.h5')
```

```python
print(model.predict([0.99]))
```

```python
plt.plot(fit_hist.history['loss'])
plt.show()
```

```python
noise = np.array(np.random.normal(0, 0.05, 100))  # 평균은 0, 표준편차는 0.05, 자료는 100개
print(noise)
```

```python
noised_scaled_data_F = np.array([])
for data in scaled_data_F:
    noised_scaled_data_F = np.append(
        noised_scaled_data_F, np.random.normal(0, 0.05, 100) + data)
print(len(noised_scaled_data_F))
print(noised_scaled_data_F[:100])
```

```python
noised_scaled_data_C = []  # 빈 리스트를 만들어두고
for data in range(0, 100):  # 노이즈 섞지 않고 그냥 데이터 주기
    for i in range(100):
        noised_scaled_data_C.append(data/100)
noised_scaled_data_C = np.array(noised_scaled_data_C)
print(len(noised_scaled_data_C))
print(noised_scaled_data_C[:101])  # 입력데이터 만개, 출력데이터 만개 (몬소리고..)
```

```python
plt.scatter(x=noised_scaled_data_C,
            y=noised_scaled_data_F, alpha=0.1)
plt.show()
```

```python
fig = plt.figure(figsize=(15, 15))
ax = fig.add_subplot()
ax.scatter(x=noised_scaled_data_C,
           y=noised_scaled_data_F,
           alpha=0.2, s=200, marker='+')  # 사이즈는 200

# x값 1개 당 y값이 100개씩, x값이 100개니까 총 만개 ㅎ
```

```python
model2 = Sequential()
model2.add(InputLayer(input_shape=(1,)))
model2.add(Dense(1))
model2.compile(loss='mse', optimizer='rmsprop')
model2.summary()
```

```python
print(model2.predict([0.01]))  # 학습 전이라 출력값 중구난방임 공부하러가자 애기야
```

```python
fit_hist = model2.fit(noised_scaled_data_C,
                      noised_scaled_data_F,
                      epochs=30)  # 데이터가 많으면 학습 20번만 해도 괜춘
```

```python
model2.save('noised_after_learning.h5')  # 학습된 애기 저장
```
